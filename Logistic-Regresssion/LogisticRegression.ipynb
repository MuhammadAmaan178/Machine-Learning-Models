```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
plt.scatter(df['age'],df['bought_insurance'], color = 'red', marker = '+')
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df['bought_insurance'],test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test)
```

```python
model.predict([[56]])
```

```python
model = LogisticRegression()
X = df[['age']]
y = df['bought_insurance']
model.fit(X,y)

age_range = np.linspace(df['age'].min(), df['age'].max(), 300).reshape(-1, 1)
predicted_probs = model.predict_proba(age_range)[:, 1]

plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['bought_insurance'], color='blue', marker='+')
plt.plot(age_range, predicted_probs, color='red', linewidth=2)
plt.xlabel('Age')
plt.ylabel('Probability of Buying Insurance')
plt.title('Logistic Regression Model')
plt.grid(True)
plt.show()
```
